```python
import requests
from bs4 import BeautifulSoup

res = requests.get('https://sports.news.naver.com/news.nhn?oid=347&aid=0000145380')
# bs4는 쓰지도 않았고 requests 라이브러리의 기능 중 get으로 해당 주소의 소스를 긁어온 것
# bs4는 Parsing용 라이브러리고 거기서 Parsing 담당 함수가 BeautifulSoup 이다.
```

### Parsing 의 개념부터 생각해야함

```python
soup = BeautifulSoup(res.content, 'html.parser') #soup은 변수처럼 보이기도 하나 실제로는 객체
```

```python
title = soup.find('title')  # 앞의 title은 파싱된 정보를 담는 객체, 뒤는 파싱된 내용 중 html title를 찾는 내용
title.get_text()
```

# 크롤링 정리
1. 크롤링할 html 코드를 가져온다.(requests.get 함수로)
2. 파싱한다. (BeautifulSoup(res.content, 'html-parser')
3. 특정 데이터를 선택한다.
4. 출력한다. (get_text())


```python
import requests
from bs4 import BeautifulSoup
# 아래 역슬레시는 여러줄에 쓸 문장
# 주소를 통해서 불러올 수도 있지만 이렇게 많은 자료를 넣고 여기서 긁어내는 방법도 있는걸 보여주는 예제

res = " <html> \
            <body> \
                <h1 id='title'>[1]크롤링이란?</h1> \
                <p class='cssstyle'>웹페이지에서 필요한 데이터를 추출하는 것</p> \
                <p id='body' align='center'>파이썬을 중심으로 다양한 웹크롤링 기술 발달</p> \
            </body> \
       </html> "

soup = BeautifulSoup(res, 'html.parser')
data = soup.find('h1')
print(data.get_text())

```

```python
import requests
from bs4 import BeautifulSoup

res = requests.get('http://v.media.daum.net/v/20170615203441266')
soup = BeautifulSoup(res.content, 'html.parser')
data = soup.find_all('span', 'txt_info')
for item in data:
    print(item.get_text())
```
